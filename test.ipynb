```python
import warnings
from tqdm import TqdmWarning
warnings.filterwarnings("ignore", category=TqdmWarning)

from huggingface_hub import snapshot_download

REPO_ID = "google/paligemma-3b-pt-224"
LOCAL_DIR = "paligemma-3b-pt-224"

snapshot_download(repo_id=REPO_ID, local_dir=LOCAL_DIR)
```

```python
import glob
from safetensors.torch import load_file

state_dict = {}
for shard in sorted(glob.glob(f"{LOCAL_DIR}/**/*.safetensors", recursive=True)):
    state_dict.update(load_file(shard))

print(f"\nTotal tensors loaded: {len(state_dict)}\n")

for name, tensor in state_dict.items():
    print(f"{name:80s} {tuple(tensor.shape)}")
```
